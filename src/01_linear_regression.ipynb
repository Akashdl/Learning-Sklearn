```python
import numpy as np
import pandas as pd
import seaborn as sns
#import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, metrics, model_selection
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X.shape
```

```python
diabetes_y.shape
```

```python
sns.scatterplot(x=diabetes_X[:,4], y=diabetes_y)
```

```python
linear_regressor = linear_model.LinearRegression()
```

```python
diabetes_X_extracted = diabetes_X[:, 2]
linear_regressor.fit(diabetes_X_extracted, diabetes_y)
```

```python
linear_regressor.fit(diabetes_X_extracted.reshape(-1,1), diabetes_y)
```

```python
# _ after coef is to signify that it is a trained value
slope_of_linear_regression_line = linear_regressor.coef_  # slope m in (m*x + b)
slope_of_linear_regression_line
```

```python
# _ after intercept is to signify that it is a trained value
intercept_of_linear_regression_line = linear_regressor.intercept_  # y-intercept b in (m*x + b)
intercept_of_linear_regression_line
```

```python
x_eval = np.linspace(-0.1, 0.2, 100)
```

```python
y_eval = linear_regressor.predict(x_eval.reshape(-1, 1))
```

```python
sns.scatterplot(x=diabetes_X_extracted , y=diabetes_y)
sns.lineplot(x=x_eval, y=y_eval)
```

# In ML dataset is divided into training data and testing data

```python
X_train = diabetes_X_extracted[:-50]
X_test = diabetes_X_extracted[-50:]
```

```python
y_train = diabetes_y[:-50]
y_test = diabetes_y[-50:]
```

```python
linear_regressor.fit(X_train.reshape(-1,1), y_train)
```

```python
y_test_pred = linear_regressor.predict(X_test.reshape(-1, 1))
```

```python
sns.scatterplot(x=X_train , y=y_train) # training dataset
sns.scatterplot(x=X_test, y=y_test) # test dataset
sns.lineplot(x=X_test, y=y_test_pred) # predicted linear regression
#sns.lineplot(x=x_eval, y=y_eval)
```

# Metrics

```python
mse = metrics.mean_squared_error(y_test, y_test_pred)
mse
```

```python
r2 = metrics.r2_score(y_test, y_test_pred)
r2
```

# train_test_split in Sklearn to select random data from original dataset for train and test datasets

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(diabetes_X_extracted, diabetes_y, test_size=0.3)
X_train.shape
```

```python
linear_regressor.fit(X_train.reshape(-1,1), y_train)
y_test_pred = linear_regressor.predict(X_test.reshape(-1, 1))
```

```python
sns.scatterplot(x=X_train , y=y_train) # training dataset
sns.scatterplot(x=X_test, y=y_test) # test dataset
sns.lineplot(x=X_test, y=y_test_pred) # predicted linear regression
```

```python
mse = metrics.mean_squared_error(y_test, y_test_pred)
r2 = metrics.r2_score(y_test, y_test_pred)
print(f"mse is {mse}")
print(f"r2 is {r2}")
```
